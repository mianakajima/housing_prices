```python
import os
#os.chdir('..')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
import numpy as np
```

```python
#read in dataset
training_set = pd.read_csv('data/train.csv')
```

```python
training_set.head()
```

```python
training_set.info()
```

```python
training_set.describe()
```

```python
#plot histograms of all variables
%matplotlib inline
training_set.hist(bins=30, figsize = (20, 15))
plt.show()
```

```python
training_copy = training_set.copy()
```

```python
# correlations
corr_matrix = training_copy.corr()
fix, ax = plt.subplots(figsize = (15, 10))
sns.heatmap(corr_matrix, 
            xticklabels = corr_matrix.columns, 
            yticklabels = corr_matrix.columns)
```

Top 10 variables with highest correlation with the sale price: 


```python
top_features = corr_matrix[['SalePrice']].sort_values('SalePrice', ascending = False)[1:11]
top_features
```

```python
matrix_columns = ['SalePrice'] + training_copy.columns[1:25].tolist()
scatter_matrix(training_copy[matrix_columns], figsize = (15, 10));
```
